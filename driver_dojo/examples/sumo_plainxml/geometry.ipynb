```python
from pyclothoids import Clothoid
import numpy as np
import matplotlib.pyplot as plt
```

```python
x1, y1 = 0.0, 0.0
t1 = np.pi / 2.0
x2, y2 = -10.0, 20.0
t2 = np.pi#np.pi / 4
cloth = Clothoid.G1Hermite(x1, y1, t1, x2, y2, t2)
plt.plot(*cloth.SampleXY(500))
plt.show()
```

```python
import pyclothoids
cloth = pyclothoids.SolveG2(x1, y1, t1, 0.0, x2, y2, t2, 1.0)
X, Y = [], []
for c in cloth:
    X += c.SampleXY(500)[0]
    Y += c.SampleXY(500)[1]
plt.plot(X, Y)
plt.show()
```
